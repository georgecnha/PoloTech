```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# sns.load_dataset('titanic')
```

### Questão 2

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Calcule a frequência relativa, relativa percentual e acumulativa para os sobreviventes no _Titanic_.

__Dica.:__ Utilize a função `.value_counts()`

```python
f_acumulada = (f_ab[0]+ f_ab[1])
print(f'A frequência acumulada é: {f_acumulada}')

f_rel = f_ab[1]/f_acumulada
print(f'A frequência relativa é: {round(f_rel,2)}')


f_rel_percent = f_rel * 100
print(f'A frequência relativa percentual é: {round(f_rel_percent,2)}%')
```

### Questão 3

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Utilizando a coluna `Age` do _dataset Titanic_, defina as principais métricas estatísticas para essa variável tais como:
- Média;
- Desvio Padrão;
- Mínimo;
- Primeiro Quartil;
- Segundo Quartil (Mediana);
- Terceiro Quartil;
- Distância Interquartil (IQR);
- Máximo;
- Skewness;
- Moda.

```python
idade = dataset['age']
```

```python
print(f'A média da idade dos passageiros era: {idade.mean()} anos')
print(f'A moda da idade dos passageiros era: {idade.mode()[0]} anos')
print(f'O Desvio Padrão da idade dos passageiros era: {idade.std()}')
print(f'O valor mínimo da idade dos passageiros era: {idade.min()} anos.')
print(f'O valor máximo da idade dos passageiros era: {idade.max()} anos.')
print(f'O Q1 da idade dos passageiros era: {idade.quantile(0.25)} anos.')
print(f'O Q2 da idade dos passageiros era: {idade.quantile(0.50)} anos.')
print(f'O Q3 da idade dos passageiros era: {idade.quantile(0.75)} anos.')
print(f'A amplitude interquartil da idade dos passageiros era: {idade.quantile(0.75)-idade.quantile(0.25)} anos.')
print(f'A assimetria (\"skewness\") da idade dos passageiros era: {idade.skew()}')
```

### Questão 4

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Crie um gráfico da dispersão das idades dos passageiros do _Titanic_ e identifique os pontos onde se encontram a média, mediana e moda das idades.

```python
sns.scatterplot(data=idade)
y_mean = idade.mean()
y_median = idade.median()
y_mode = idade.mode()[0]
plt.axhline(y_mean, color='r', linestyle='--',label='Média de idade')
plt.axhline(y_median, color='g', linestyle='--',label='Mediana da idade')
plt.axhline(y_mode, color='b', linestyle='--',label='Moda da idade')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.title('Dispersão das idades dos passageiros')
plt.show()



#Estamos definindo as coordenadas (1.05, 1) para o parâmetro bbox_to_anchor, que significa que a legenda 
#será posicionada 5% além da largura do gráfico no eixo x (1.05) e no topo da altura do gráfico no eixo y. 
```

```python
# Versão da resulução da questão com histograma

plt.figure(figsize=(12, 6))
sns.histplot(dataset["age"], kde=True)
y_mean = idade.mean()
y_median = idade.median()
y_mode = idade.mode()[0]
plt.axvline(y_mean, color='r', linestyle='--',label='Média de idade')
plt.axvline(y_median, color='g', linestyle='--',label='Mediana da idade')
plt.axvline(y_mode, color='b', linestyle='--',label='Moda da idade')
plt.legend(bbox_to_anchor=(1.21, 1))
plt.title('Dispersão das idades dos passageiros', size=15)
plt.show()



#Estamos definindo as coordenadas (1.05, 1) para o parâmetro bbox_to_anchor, que significa que a legenda 
#será posicionada 5% além da largura do gráfico no eixo x (1.05) e no topo da altura do gráfico no eixo y. 
```
