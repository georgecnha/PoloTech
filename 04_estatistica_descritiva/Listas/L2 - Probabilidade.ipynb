A probabilidade de dado honesto é conhecida e igual a $1/6$, mas quando vamos testar isso com um dado físico, não necessariamente vai cair uma vez cada face do dado (se eu lançar ele 6 vezes). A ideia é que conforme fazermos vários testes a tendência é que a probabilidade de cada face irá convergir pro valor esperado. 

Vamos testar este conceito utilizando da programação seguintes os itens abaixo:]

__A)__ Desenvolva uma função que receba um valor n, onde n é a quantidade de elementos e o objetivo da função é retorna uma _Series_ com n elementos sendo esses elementos números variando de 1 e 6;

__B)__ Teste a função para n igual a 10, 100, 1.000, 10.000 e 100.000 casos e para cada um deles calcule a proporção de vezes que aparece cada número;

__DIca.:__ utilize a função do _Pandas_ `.value_counts`

```python
# A

import pandas as pd
import numpy as np
from seaborn import histplot

def dado(n):
    a = np.random.randint(1,7,n)
    return (pd.Series(a))
```

```python
# B 

#N =  10
n_10= dado(10)
print('Para n = 10')
print(n_10.value_counts())
histplot(n_10, discrete = True, bins = 6);
```

- 14 balas azuis;
- 22 balas vermelhas;
- 17 balas amarelas.

A mamãe deles deixou que cada um pegasse 3 balas começando pela Luiza e depois o Rafael. Dado isso resolva os item abaixo:

__A)__ Qual a probabilidade da Luiza pegar uma bala de cada na ordem da contagem e o Rafael pegar 2 azuis e uma vermelha, dado que é um evento sem reposição?

__B)__ Luiza decidiu ficar só com a bala amarela e devolveu as demais para o pote antes que o Rafael pegasse suas balas. Como que ficaria a probabilidade para o Rafael agora?

```python
# Dados
b = 14
r = 22
y = 17

balas = [b,r,y]
total = sum(balas)
```

```python
# A)
# Luiza

p_b = b/total
total-=1 
b-=1


# Eliminando a retirada do total

p_r = r/(total)
total-=1
r-=1

p_y = y/(total)
y-=1
total-=1

p_bry = (p_b*p_r*p_y)

print(f'A chance de Luiza pegar cada bala na ordem da contagem é de {p_bry*100:.3f}%.')
```

### Questão 4

Um programa de computador gera uma pontuação aleatória para cada usuário, sendo que:

- Em 1/2 dos casos, gera 0.5;
- Em 1/4 dos casos, gera 0.25;
- Em 1/8 dos casos, gera 0.125.

E essa lógica segue assim por diante. Resolva os itens a seguir:

__A)__ Calcule o __valor esperado__ utilizando apenas os 3 primeiros casos;

__B)__ Faça um _loop_ para calcular o __valor esperado__ para 100 casos e avalie se só com os 3 primeiros casos temos uma boa aproximação ou não.

### Anotações
#### Esperança

- Discreta:

$E(x) = x_1 \cdot p(x_1) +  x_2 \cdot p(x_2) +  x_3 \cdot p(x_3)... = \sum\limits_{i=1}^n {x_i \cdot p(x_i)}$

- Contínua:

$E(x)=\int\limits_1^{\infty} xf(x)dx$



```python
# A) 

esperanca_a = (1/2)*0.5 + (1/4)*.25 + (1/8)*.125

print(esperanca_a)
```

B)
### Anotações

$ E(x) = \dfrac{1}{2} \cdot \dfrac{1}{2} +  \dfrac{1}{4} \cdot \dfrac{1}{4} +  \dfrac{1}{8} \cdot \dfrac{1}{8}$ 

$ E(x) = \dfrac{1}{4} +  \dfrac{1}{16} +  \dfrac{1}{64} $

$ E(x) = (\frac{1}{2})^2 +  (\frac{1}{4})^2 +  (\frac{1}{8})^2 $

$ E(x) = (\frac{1}{2^1})^2 + (\frac{1}{2^2})^2  + (\frac{1}{2^3})^2$

Aproximo a esperança com a seguinte equação:

$ E(x) = \sum\limits_1^{n}  \left(\dfrac{1}{2^n}\right)^2  $

```python
# B)
n = 100
esperanca_b1 = 0

for i in range(1,n+1):
    esp_parcial = (1/(2**i)) **2
    esperanca_b1 += esp_parcial
    
print(f'A esperança para {n} números é: {esperanca_b1}.')
    
print(f'E(a)/E(b) = {esperanca_a/esperanca_b1}')
```

```python
# Se n = 3 

n = 3
esperanca_b2 = 0

for i in range(1,n+1):
    esp_parcial = (1/(2**i)) **2
    esperanca_b2 += esp_parcial
    
print(f'A esperança para {n} números é: {esperanca_b2}.')
print(f'E(a)/E(b) = {esperanca_a/esperanca_b2}')    
```
